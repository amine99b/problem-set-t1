```python
import pandas as pd
import numpy as np
```

```python
file = 'ccd_image.txt'
```

```python
with open(file, 'r') as file:
    content = file.read()
print(content)
```

```python
import re
```

```python
numbers_str = re.findall(r'\d+', content)
numbers = [int(x) for x in numbers_str]
mean = sum(numbers) / len(numbers)
print(f"Mean: {mean}")
```

```python
import statistics
```

```python
std_dev = statistics.stdev(numbers)
print(f"Standard Deviation: {std_dev}")
```

```python
median = statistics.median(numbers)
print(f"Median: {median}")
```

```python
mode = statistics.mode(numbers)
print(f"Mode: {mode}")
```

```python
import matplotlib.pyplot as plt
```

```python
range_min = 1000
range_max = 3000
num_bins = 100
```

```python
hist, bins = np.histogram(numbers, bins=num_bins, range=(range_min, range_max))
log_hist = np.log10(hist)
plt.figure(figsize=(10, 5))
plt.bar(bins[:-1], log_hist, width=(range_max - range_min) / num_bins, align='edge')
plt.xlabel('ADU (Analog-to-Digital Units)')
plt.ylabel('Logarithmic Frequency')
plt.title('Logarithmic Histogram of Data')
plt.show()
```

```python
#the median would be the most robust and suitable statistic to estimate the sky level. 
#Because of it's Robustness to Outliers.
#provides a better estimate of the central value in skewed distributions such as sky level.
```
